```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("Churn.csv")
```

```python
df
```

```python
# Calculate the IQR for each column
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
```

```python
df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
```
